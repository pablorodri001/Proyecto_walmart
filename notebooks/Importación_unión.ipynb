```python
import pandas as pd
```

```python
#probamos a crear un dataframe con pandas y a leer el archivo train.csv
df_train=pd.read_csv("../data/raw/train.csv")
df_train.head
```

```python
df_stores=pd.read_csv("../data/raw/stores.csv")
df_stores.head
```

```python
df_features=pd.read_csv("../data/raw/features.csv")
df_features.head
```

```python
duplicados = df_train.duplicated(subset=['Store', 'Dept', 'Date']).sum()
if duplicados > 0:
    print("Mostrando registros duplicados:")
    print(df_train[df_train.duplicated(subset=['Store', 'Dept', 'Date'], keep=False)])
```

```python
lista_Dataframes=[df_train,df_features]
for df in lista_Dataframes:
    df['Date'] = pd.to_datetime(df['Date'])

df_train.head()
```

```python
datos=df_train.merge(df_features, on=['Store','Date'], how='left')
datos=datos.merge(df_stores, on='Store',how='left')
```

```python
datos.head()
```

```python
print(f"Filas originales en train: {df_train.shape[0]}")
print(f"Filas en el dataset unido: {datos.shape[0]}")
```

```python
datos.columns
```

```python
datos=datos.drop(columns="IsHoliday_y")
datos.columns
```

```python
datos=datos.rename(columns={'IsHoliday_x' : 'IsHoliday'})
```

```python
datos.to_csv("../data/processed/datos.csv",index='False')
```
