```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../data/processed/datos.csv',parse_dates=['Date'])
```

```python
print(df.shape)
df.head()
```

```python
df_agrupado_tipo=df.groupby('Type')
df_agrupado_tipo.head()
```

```python
df_media_tipo=df_agrupado_tipo['Weekly_Sales'].mean().sort_values(ascending=False).reset_index()
df_media_tipo
#Observamos que las tiendas tipo A, venden más semanalmente, averiguaremos si tiene correlación con la superficie de la tienda
```

```python
df_media_tamaño=df_agrupado_tipo['Size'].mean().reset_index()
df_media_tamaño.head()
#Obeservamos que las tiendas A son mucho más grandes que las B y las C así que podemos correlacionar que las ventas dependen del tamaño de la tienda
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 6))
sns.barplot(x='Type', y='Weekly_Sales',data=df_media_tipo, ax=ax[0], palette='magma')
ax[0].set_title('Ventas por tipo de tienda')
sns.barplot(x='Type', y='Size',data=df_media_tamaño, ax=ax[1], palette='viridis')
ax[1].set_title('tamaño por tipo')
plt.tight_layout()
plt.show()
```

```python
df['Month']=df['Date'].dt.month
df['Year']=df['Date'].dt.year
df
```

```python
df_agrupación_festivos=df.groupby('IsHoliday')
df_agrupación_festivos = df_agrupación_festivos['Weekly_Sales'].mean().reset_index()
df_agrupación_festivos
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='IsHoliday',y='Weekly_Sales', data=df_agrupación_festivos, palette='Set2')
plt.title("ventas_festivos")
plt.show()
```

```python
df_agrupación_mes=df.groupby('Month')
df_agrupación_mes=df_agrupación_mes['Weekly_Sales'].mean().reset_index()
df_agrupación_mes
```

```python
plt.figure(figsize=(5,8))
sns.lineplot(x='Month', y='Weekly_Sales',data=df_agrupación_mes, marker='o',color='red', linewidth=2.5)
plt.title('Tendencia de Ventas a lo largo del Año (Promedio)')
plt.xticks(range(1, 13))
plt.grid(True, alpha=0.3)
plt.show()
```

```python
columnas_de_Interes=['Weekly_Sales','Temperature','Fuel_Price','CPI','Unemployment','Size']
df_corr=df[columnas_de_Interes].corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df_corr,annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```
