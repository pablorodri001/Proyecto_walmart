```python
import pandas as pd
```

```python
df=pd.read_csv('../data/processed/datos.csv',parse_dates=['Date'])
```

```python
print(df.shape)
df.head()
```

```python
df_agrupado_tipo=df.groupby('Type')
df_agrupado_tipo.head()
```

```python
df_media_tipo=df_agrupado_tipo['Weekly_Sales'].mean().sort_values(ascending=False)
df_media_tipo
#Observamos que las tiendas tipo A, venden más semanalmente, averiguaremos si tiene correlación con la superficie de la tienda
```

```python
df_media_tamaño=df_agrupado_tipo['Size'].mean()
df_media_tamaño.head()
#Obeservamos que las tiendas A son mucho más grandes que las B y las C así que podemos correlacionar que las ventas dependen del tamaño de la tienda
```
